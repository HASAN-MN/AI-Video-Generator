```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils  import export_to_video
#users diffusers to use diffusion and in the context of generative  models, generating videos from noise (guassian noise)
#from that noise we are trying to create videos in the output, working alongside the model.
pipe= DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b",torch_dtype=torch.float16,
                                        variant= "fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()

#creating a pipeline with all of this along with the vlab pretrained model i.e installed.
```

```python
prompt=input("Provide text to convert into video: ")
video_frames=pipe(prompt,num_inference_steps=35).frames
vid_path= export_to_video(video_frames)
vid_name=vid_path.replace('/tmp','')
print('Name: ', vid_name)
torch.cuda.empty_cache()
```
